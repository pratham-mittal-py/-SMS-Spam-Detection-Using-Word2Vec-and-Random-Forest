```python
import gensim
from gensim.models import Word2Vec
```

```python
import pandas as pd
messages=pd.read_csv('SMSSpamCollection.txt',
                    sep='\t',names=["label","message"])
```

```python
! pip install nltk
import re
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer() 
```

```python
corpus=[]
for i in range (0, len(messages)):
    review= re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review=review.lower().split()
    review= [lemmatizer.lemmatize(word) for word in review]
    corpus.append (' '. join(review))
```

```python
corpus
```

```python
from nltk import sent_tokenize, word_tokenize
from gensim.utils import simple_preprocess

```

```python
words_of_each_sentence_by_sentence=[]
for msg in corpus:
    sent_token=sent_tokenize(msg)
    for sent in sent_token:
        word=simple_preprocess(sent)
        words_of_each_sentence_by_sentence.append(word)
```

# Implementing Avg Word2Vec

```python
import numpy as np
def avg_word2vec(doc):
    vectors = [model.wv[word] for word in doc if word in model.wv.index_to_key]
    if len(vectors) == 0:
        # Return zero vector of correct size if no words are known
        return np.zeros(model.vector_size)
    return np.mean(vectors, axis=0)
```

```python
def preprocess_input(text):
    text = re.sub('[^a-zA-Z]', ' ', text)
    text = text.lower().split()
    text = [lemmatizer.lemmatize(word) for word in text]
    tokens = simple_preprocess(' '.join(text))
    return tokens
```

```python
def predict_message(message):
    tokens = preprocess_input(message)
    vec = avg_word2vec(tokens).reshape(1, -1)
    prediction = classifier.predict(vec)[0]
    return "ham" if prediction == 1 else "spam"
```

```python
user_input = input("Enter an SMS message: ")
result = predict_message(user_input)
print(f"Prediction: {result}")
```
